```python
def bigram(words): # key = (wi-1, wi)      value =  P(wi | wi-1) 
    
    # Count(wi-1)
    count_wmt = {}
    for w in words:
        count_wmt[w] = count_wmt.get(w, 0) + 1

    # Count(wi-1, wi)
    count_wim1_wi = {}
    for i in range(1, len(words)):
        key = (words[i - 1], words[i])
        count_wim1_wi[key] = count_wim1_wi.get(key, 0) + 1

    # calc P(wi | wi-1) = Count(wi-1, wi) / Count(wi-1)
    p={}
    for k, v in count_wim1_wi.items():
        key = k[0]
        p[k] = v / count_wmt[key]
    
    return p

bigram(words)
```

```python
def trigram(words): # key = (wi-2, wi-1, wi)      value = P(wi | wi-2, wi-1)

    # Count(wi-2, wi-1)
    count_wim2_wim1 = {}
    for i in range(1, len(words)):
        key = (words[i - 1], words[i])
        count_wim2_wim1[key] = count_wim2_wim1.get(key, 0) + 1
    
    # Count(wi-2, wi-1, wi)
    count_wim2_wim1_wi = {}
    for i in range(2, len(words)):
        key = (words[i - 2], words[i - 1], words[i])
        count_wim2_wim1_wi[key] = count_wim2_wim1_wi.get(key, 0) + 1

    # calc P(wi | wi-2, wi-1) = Count(wi-2, wi-1, wi) / Count(wi-2, wi-1)
    p={}
    for k, v in count_wim2_wim1_wi.items():
        key = (k[0], k[1])
        p[k] = v / count_wim2_wim1[key]
    
    return p

trigram(words)
```
